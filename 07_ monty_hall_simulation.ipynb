```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 10000 # number of trials
print_debug = False
plot_results = True

stay_wins = 0
switch_wins = 0
stay_series = []
switch_series = []
choices = [1,2,3]

for i in range(N):
    player_choice, correct_door = np.random.choice(choices, size=2)
    rem_choices = set(choices) - set([player_choice, correct_door])
    monty_choice = np.random.choice(list(rem_choices))    
    switch_choice = list(set(choices) - set([player_choice, monty_choice]))[0]
    if player_choice == correct_door:
        stay_wins += 1
    elif switch_choice == correct_door:
        switch_wins += 1
    stay_series.append(stay_wins)
    switch_series.append(switch_wins)
    if print_debug:
        print("Iter %d, player choice = %d, monty_choice = %d, switch_choice = %d, correct door = %d, stay_wins = %d, switch_wins = %d" % (i, player_choice, monty_choice, switch_choice, correct_door, stay_wins, switch_wins))

switch_winprob = switch_wins / (switch_wins + stay_wins)
stay_winprob = stay_wins / (stay_wins + switch_wins)
print("Switch win prob = %.2f Stay win prob = %.2f" % (switch_winprob, stay_winprob))
        
if plot_results:
    x = range(1, N + 1)
    plt.plot(x, stay_series, label='Stay wins')
    plt.plot(x, switch_series, label='Switch wins')
    plt.xlabel('Trials')
    plt.ylabel('Cumulative wins')
    plt.legend()
```
